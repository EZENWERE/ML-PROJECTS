```python
import pandas as pd 
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins= 10);
```

```python
heart_disease.plot.hist(figsize = (10,30), subplots = True, bins =20 );

```

* when plotting something quicl;y, okay to use the plplot method
* when plotting something more advanced, use the oo method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50] #get the actual over 50 dataframe
len(over_50)
```

```python
over_50
```

```python
#pyplot method
over_50.plot(kind ="scatter", x = "age", y = "chol", c="target", figsize =(10, 6));
```

```python
#OO method mixed with pyplot method
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot( ax= ax, kind ="scatter", x = "age", y = "chol", c= "target")
ax.set_xlim([45, 100]); # has an x range limit 
```

```python
# oo method from scratch 
fig, ax = plt.subplots(figsize = (10, 6))

# plot the data
scatter = ax.scatter(x= over_50["age"],
                     y=over_50["chol"],
                     c =over_50["target"] ) # legend_element!
#CUSTOMIZE THE PLOT 
ax.set(title ="Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "cholesterol"
      );

# Add legend
ax.legend(*scatter.legend_elements(), title ="Target");

# Add a horizontal line 
ax.axhline(over_50["chol"].mean(), ls = ':');
```

```python
over_50.target.values
```

```python
over_50.head()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows =2 ,
                               ncols=1
                              , figsize=(5,10),
                               sharex =True) #share the x axis
#Add data to ax1 amd ax2
scatter = ax1.scatter(x= over_50["age"],
                     y=over_50["chol"],
                     c =over_50["target"] ) # legend_element!

scatter = ax2.scatter(x= over_50["age"],
                     y=over_50["thalach"],
                     c =over_50["target"] ) # legend_element!
#CUSTOMIZE THE PLOT 
ax1.set(title ="Heart Disease and Cholesterol Levels",
       ylabel = "cholesterol"
      );
ax2.set(title ="Heart Disease and Max heart rate",
       xlabel = "Age",
       ylabel = "Max heart rate"
      );

# Add legend
ax1.legend(*scatter.legend_elements(), title ="Target");#gets legend elements and gives it a title 
#on the figure
ax2.legend(*scatter.legend_elements(), title ="Target");

# Add a horizontal line 

ax1.axhline(over_50["chol"].mean(), ls = ':');
ax2.axhline(over_50["thalach"].mean(), ls = "--");

#figure title
fig.suptitle("Heart disease Analysis", fontsize =16, fontweight ="bold");
```

## Customizing an inbuilt style

```python

plt.style.use('seaborn-v0_8-whitegrid')
# oo method from scratch 
fig, ax = plt.subplots(figsize = (10, 6))

# plot the data
scatter = ax.scatter(x= over_50["age"],
                     y=over_50["chol"],
                     c =over_50["target"],  # legend_element!
                     cmap="plasma") #changes colour scheme
#CUSTOMIZE THE PLOT 
ax.set(title ="Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "cholesterol"
      );

# Add legend
ax.legend(*scatter.legend_elements(), title ="Target");

# Add a horizontal line 
ax.axhline(over_50["chol"].mean(), ls = ':');
```

```python
# Customizing the y and x axis limitations 
fig, (ax1, ax2) = plt.subplots(nrows =2 ,
                               ncols=1
                              , figsize=(5,10),
                               sharex =True) #share the x axis
#Add data to ax1 amd ax2
scatter = ax1.scatter(x= over_50["age"],
                     y=over_50["chol"],
                     c =over_50["target"], # legend_element!
                     cmap ="winter") 

scatter = ax2.scatter(x= over_50["age"],
                     y=over_50["thalach"],
                     c =over_50["target"], # legend_element!
                     cmap= "summer") 
#CUSTOMIZE THE PLOT (using x_lim and y_lim to remove the unecessary axis line plotted in original figure )
ax1.set(title ="Heart Disease and Cholesterol Levels",
       ylabel = "cholesterol"
      );
ax1.set_xlim([50,80])

ax2.set(title ="Heart Disease and Max heart rate",
       xlabel = "Age",
       ylabel = "Max heart rate"
      );
ax1.set_xlim([50, 80])
ax2.set_ylim([60,200])


# Add legend
ax1.legend(*scatter.legend_elements(), title ="Target");
ax2.legend(*scatter.legend_elements(), title ="Target");

# Add a horizontal meanline 
ax1.axhline(over_50["chol"].mean(), ls = ':');
ax2.axhline(over_50["thalach"].mean(), ls = "--");

#figure title
fig.suptitle("Heart disease Analysis", fontsize =16, fontweight ="bold");

```
